```javascript
function fizzbuzz(n) {
  if (n > 1) {
    fizzbuzz(n - 1);
  }

  if (n % 3 == 0 && n % 5 == 0) {
    console.log("Fizz,Buzz");
  }
  else if (n % 3 == 0) {
    console.log("Fizz");
  }
  else if (n % 5 == 0) {
    console.log("Buzz");
  }
  else {
    console.log(n);
  }
};
```

```javascript
function nabeatsu(i) {
    if (i > 1) {
        nabeatsu(i - 1);
    }

    let text = i.toString();

    if (i % 3 == 0) {
        console.log(text + "!!!");
    }
    else if (text.indexOf("3") !== -1) {
        console.log(text + "!!!");
    }
    else {
        console.log(i);
    }
};
```

```javascript
fizzbuzz(10000);
```

```javascript
nabeatsu(1000);
```
